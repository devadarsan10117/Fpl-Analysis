```python
file_path = '/Users/devadarsanna/Downloads/fpl_player_statistics.csv'

# Load the CSV
df = pd.read_csv(file_path)

# Show the first few rows
print(df.head())
```

```python
# 2Ô∏è‚É£ List of columns to remove
cols_to_remove = [
    # Event / Gameweek Specific
    'event_points', 'transfers_in_event', 'transfers_out_event',
    'cost_change_event', 'cost_change_event_fall',

    # Change Tracking
    'cost_change_start', 'cost_change_start_fall',

    # Internal FPL Metrics
    'influence', 'creativity', 'threat',
    'influence_rank', 'creativity_rank', 'threat_rank', 'ict_index_rank',
    'influence_rank_type', 'creativity_rank_type', 'threat_rank_type', 'ict_index_rank_type',

    # Per 90 Metrics
    'expected_goals_per_90', 'expected_assists_per_90',
    'expected_goal_involvements_per_90', 'expected_goals_conceded_per_90',
    'goals_conceded_per_90', 'saves_per_90', 'clean_sheets_per_90', 'starts_per_90',

    # Set Piece Info
    'corners_and_indirect_freekicks_order', 'corners_and_indirect_freekicks_text',
    'direct_freekicks_order', 'direct_freekicks_text',
    'penalties_order', 'penalties_text',

    # Discipline / Rare Events
    'yellow_cards', 'red_cards', 'own_goals',
    'penalties_saved', 'penalties_missed',

    # Defensive Detail
    'tackles', 'clearances_blocks_interceptions',
    'recoveries', 'defensive_contribution', 'defensive_contribution_per_90',

    # Transfer Totals
    'transfers_in', 'transfers_out',

    # Miscellaneous / Redundant
    'cost_change_start_fall', 'cost_change_event_fall'
]

# 3Ô∏è‚É£ Drop the columns (safely ‚Äî ignores if a column doesn‚Äôt exist)
df = df.drop(columns=[col for col in cols_to_remove if col in df.columns])

print("‚úÖ Unnecessary columns removed!")
print("New shape:", df.shape)

# 4Ô∏è‚É£ Save cleaned dataset
clean_path = '/Users/devadarsanna/Downloads/fpl_player_statistics_cleaned.csv'
df.to_csv(clean_path, index=False)

print(f"üíæ Cleaned file saved as: {clean_path}")
print("Remaining columns:")
print(df.columns.tolist())
```
